```python
import os
import pandas as pd
import re
```

```python
path = os.path.join("Data","train.csv")

df = pd.read_csv(path)
df.head()
```

```python
df.set_index(df["id"],inplace=True)
df.drop(columns = ["id"],inplace = True)
df.head()
```

```python
df["Size"].unique()
```

```python
print(df.isnull().sum())
df.dropna(how="any",inplace=True)
print(df.isnull().sum())
```

```python
def Preprocessing(data):
    data["Laptop Compartment"] = data["Laptop Compartment"].map({"Yes":1,"No":0})
    data["Size"] = data["Size"].map({"Small":0,"Medium":1,"Large":2})
    return data
df.head()
```

```python
preprocessed_data = Preprocessing(df)
preprocessed_data.head()
```
